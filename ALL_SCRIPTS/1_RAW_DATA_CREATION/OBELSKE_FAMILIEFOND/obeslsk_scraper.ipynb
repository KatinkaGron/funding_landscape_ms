```python
import pandas as pd
import numpy as np
import time
import re as re
from selenium import webdriver
import translators as ts
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver = webdriver.Chrome()
def extract_data_obelsk(url):
    
    """
    This function will collect the grant information provided from the Obelske foundation grant database with the url: https://obel.com/alle-bevillinger/
    
    The paths are found using devtools, and found through the webpage.
    
    note: When the window opens, one must actively choose a web browser (eg. Ecosia) and close the
    the Cookie popup window - to start the extraction. 
    
    The information is saved in the lists of the following groups:
    
    receiver_l = []
    description_l = []
    year_l = []
    category_l = []
    amount_L = []
    
    and if data is not avaiable for a certain group, then "NA" is inserted. 
    
    The lists are finally zipped together in the biglist dataframe.
    
    input: target_url (https://obel.com/alle-bevillinger/")
    return: Dataframe with text extracted
    """
    
    
    receiver_l = []
    description_l = []
    year_l = []
    category_l = []
    amount_L = []
    
    
    driver.get(url)
    driver.maximize_window() #opening full web browser window
    time.sleep(20)
    
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#appropriations_table > tbody > tr:nth-child(1)")))
    action = ActionChains(driver)
    action.move_to_element(element).perform()


    time.sleep(5)
    try:
        for page_num in range(1,29): #29
            """we click next page 29 times"""
            for i in range(1,26):
                time.sleep(2)
                
                action.move_to_element(driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td:nth-child(1)")).perform()  #move to the line of interest

                receiver = driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td:nth-child(1)").text #receiver

                description = driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td:nth-child(2)").text #description
                year = driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td.dt-type-numeric.sorting_1").text #year
                category = driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td:nth-child(4)").text # cagtegory
                amount = driver.find_element(By.CSS_SELECTOR, f"#appropriations_table > tbody > tr:nth-child({i}) > td.dt-type-numeric.divide").text # amount
                if i == 25:
                    print(description)
                #append
                receiver_l.append(receiver)
                description_l.append(description)
                year_l.append(year)
                category_l.append(category)
                amount_L.append(amount)


            

            ActionChains(driver).move_to_element(driver.find_element(By.CSS_SELECTOR, "#appropriations_table_wrapper > div.dt-paging.paging_full_numbers > button.dt-paging-button.next")).perform()  #move to click button
            element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#appropriations_table_wrapper > div.dt-paging.paging_full_numbers > button.dt-paging-button.next")))
            element.click()

                
            
    except NoSuchElementException: #avoid programme crashing
        print("Loop ended")

 

    df_final = pd.DataFrame(list(zip(year_l, receiver_l, amount_L, description_l, category_l)), columns=["Year",'Receiver','Grant size (DKK)', 'Description', "Project Category"])

    return df_final

biglist = extract_data_obelsk("https://obel.com/alle-bevillinger/")
pd.set_option('display.max_columns', None)

biglist.to_csv('obelsk_data.csv', index=True)
```
