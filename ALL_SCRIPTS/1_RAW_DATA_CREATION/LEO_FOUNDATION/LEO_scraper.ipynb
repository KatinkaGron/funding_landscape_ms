```python
import pandas as pd
import numpy as np
import time
import re as re
from selenium import webdriver
import translators as ts
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver = webdriver.Chrome()
def get_data_leo(url):
    
    """Leo foundation
    note: this programme only works until page 30!
    next function: get_data_leo2pages will get the two last pages 
    
        This function will collect the grant information provided from the Leo foundation foundation grant database with the url: "https://leo-foundation.org/en/grantee/"
    
    The paths are found using devtools, and found through the webpage.
    
    note: When the window opens, one must actively choose a web browser (eg. Ecosia) and close the
    the Cookie popup window - to start the extraction. 
    
    The information is saved in the lists of the following groups:
    year_l = []
    title_l = []
    project_category_l = []
    receiver_name = []
    description_l = []
    amount_L = []
    location_l = []
    
    and if data is not avaiable for a certain group, then "NA" is inserted. 
    
    The lists are finally zipped together in the biglist dataframe.
    
    input: target_url ("https://leo-foundation.org/en/grantee/")
    return: Dataframe with text extracted
    
    """

    year_l = []
    title_l = []
    project_category_l = []
    receiver_name = []
    description_l = []
    amount_L = []
    location_l = []
    
    
    driver.get(url)
    driver.maximize_window() #opening full web browser window
    time.sleep(5)
    

    action = ActionChains(driver)
    for page_num in range(1,32): #32
        time.sleep(5)
        for i in range(1,11):
            try:
                try:
                    ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/button")).perform()  #move to place
                    elements = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/button")))
                    elements.click()
                    time.sleep(5)
                except ElementClickInterceptedException:
                    print("could not click")
                    print(i)
                    continue

                ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/h3")).perform()  #move to place
                description = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[2]").text
                project_cat = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[1]").text  #project category
                year = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[2]").text  #year
                location = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[3]").text  #location
                amount = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[2]/p[2]").text #Amount
                rec_name = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[2]/p[1]").text #grantee name
                title = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/h3").text #title

                year_l.append(year)
                project_category_l.append(project_cat)
                receiver_name.append(rec_name)
                description_l.append(description)
                amount_L.append(amount)
                location_l.append(location)
                title_l.append(title)

            except NoSuchElementException: #avoid programme crashing
                year = "NA"
                project_cat = "NA"
                rec_name = "NA"
                description = "NA"
                amount = "NA"
                location = "NA"

                year_l.append(year)
                project_category_l.append(project_cat)
                receiver_name.append(rec_name)
                description_l.append(description)
                amount_L.append(amount)
                location_l.append(location)
                title_l.append(title)

                print("Loop ended")
                print(i)
        try:
            ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[3]/ul/li[{page_num+1}]/a")).perform()  #move to click button
            element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[3]/ul/li[{page_num+1}]/a")))
            element.click()
            #click

        except ElementClickInterceptedException:
            print("could not turn page")
        except NoSuchElementException:
            print("programme ended")



    df_final = pd.DataFrame(list(zip(year_l, receiver_name, amount_L, description_l, project_category_l, location_l, title_l)), columns=["Year",'Reciever','Grant size (DKK)', 'Description', "Project Category", "Location", "Title"])

    return df_final

driver = webdriver.Chrome()

def get_data_leo2pages(url):
    
    """Leo foundation
    This function is exactly the same as x except the input url is now:
    https://leo-foundation.org/en/grantee/page/31/
    
    
    """

    year_l = []
    title_l = []
    project_category_l = []
    receiver_name = []
    description_l = []
    amount_L = []
    location_l = []
    
    
    driver.get(url)
    driver.maximize_window() #opening full web browser window
    time.sleep(5)
    

    action = ActionChains(driver)
    for page_num in range(1,3): #32
        time.sleep(5)
        for i in range(1,11):
            try:
                try:
                    ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/button")).perform()  #move to place
                    elements = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/button")))
                    elements.click()
                    time.sleep(5)
                except ElementClickInterceptedException:
                    print("could not click")
                    print(i)
                    continue

                ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/h3")).perform()  #move to place

                
                description = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[2]").text
                project_cat = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[1]").text  #project category
                year = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[2]").text  #year
                location = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[2]/div[1]/p[3]").text  #location
                amount = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[2]/p[2]").text #Amount
                rec_name = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[2]/p[1]").text #grantee name
                title = driver.find_element(By.XPATH, f"/html/body/div/div[1]/main/div/div/div/main/div/div[2]/div[{i}]/div[1]/div[1]/div/h3").text #title

                year_l.append(year)
                project_category_l.append(project_cat)
                receiver_name.append(rec_name)
                description_l.append(description)
                amount_L.append(amount)
                location_l.append(location)
                title_l.append(title)

            except NoSuchElementException: #avoid programme crashing
                year = "NA"
                project_cat = "NA"
                rec_name = "NA"
                description = "NA"
                amount = "NA"
                location = "NA"

                year_l.append(year)
                project_category_l.append(project_cat)
                receiver_name.append(rec_name)
                description_l.append(description)
                amount_L.append(amount)
                location_l.append(location)
                title_l.append(title)

                print("Loop ended")
                print(i)
        try:
            ActionChains(driver).move_to_element(driver.find_element(By.XPATH, f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[3]/ul/li[7]/a")).perform()  #move to click button
            element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"/html/body/div[2]/div[1]/main/div/div/div/main/div/div[3]/ul/li[7]/a")))
            element.click()
            #click

        except ElementClickInterceptedException:
            print("could not turn page")
        except NoSuchElementException:
            print("programme ended")



    df_final2 = pd.DataFrame(list(zip(year_l, receiver_name, amount_L, description_l, project_category_l, location_l, title_l)), columns=["Year",'Reciever','Grant size (DKK)', 'Description', "Project Category", "Location", "Title"])

    return df_final2
biglist = get_data_leo("https://leo-foundation.org/en/grantee/")
pd.set_option('display.max_columns', None)


biglist_2pages = get_data_leo2pages("https://leo-foundation.org/en/grantee/page/31/")
pd.set_option('display.max_columns', None)


list_merged = pd.concat([biglist, biglist_2pages])  #if it raises an error, write to csv using with!
list_merged.to_csv('leo_data.csv', index=True)  
```
